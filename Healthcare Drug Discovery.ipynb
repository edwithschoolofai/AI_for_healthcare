## 신약 개발 과정

![alt text](https://image.slidesharecdn.com/drugdiscoveryanddevelopment-111227060152-phpapp02/95/drug-discovery-and-development-11-638.jpg?cb=1422671083 "Logo Title Text 1")

- 신약의 초기 발견부터 정식 판매까지는 수년에서 수십 년이 걸립니다.
- 각종 실험과 임상 연구, 그리고 임상시험이 필요합니다.
- 동물 실험을 통과한 신약들 중 무려 90%가 인간 대상 임상시험에서 실패합니다.

#### 1단계 - 의학 문헌 공부

![alt text](https://image.slidesharecdn.com/drugdiscoveryprocessstyle3powerpointpresentationtemplates-120316015043-phpapp02/95/drug-discovery-process-style-3-powerpoint-presentation-templates-1-728.jpg?cb=1331862866 "Logo Title Text 1")

- 의사들은 다른 연구들에서 소개된 약물과 질병, 그리고 단백질 간의 관계에 대해 공부해야 합니다. 
- 그들은 약물의 목표가 무엇인지, 예를 들면 어떤 단백질과 결합해야 하는지 등을 알아내야 합니다. 

#### 2단계 - 약물이 가져야 할 성질 조사

![alt text](https://www.scientist.com/wp-content/uploads/2013/11/Stem-cells-can-be-used-to-determine-drug-like-properties.jpg "Logo Title Text 1")

- 약물로부터 어떤 성질을 원하는가?
- 얼마나 잘 녹아야 하는가?
- 이 단백질과 결합하기 위해서는 어떤 구조를 가져야 하는가?
- 어떤 종류의 암을 치료할 것인가?

#### 3단계 - 해당 성질을 가진 분자 조사

![alt text](http://images.slideplayer.com/28/9337949/slides/slide_4.jpg "Logo Title Text 1")

- 한 데이터베이스는 7,200만 개의 분자와 그들의 구조식, 성질 등을 포함한 모든 것이 나열되어 있습니다.
- 특정 분자가 특정 질병을 치료하는지를 알아내야 합니다.

#### 4단계 - 실험!

![alt text](http://humanspareparts.fi/wp-content/uploads/2015/06/hsp_infographic_4.jpg "Logo Title Text 1")

- 리드 분자라 불리는 그들의 생각이 실험적 검증을 위해 실험실로 보내집니다.
- 만약 실험실에서 효과가 있다고 판단하면, 임상시험을 시작할 수 있습니다.
- 아주 작은 비율의 약물만이 실제로 모든 과정을 거쳐 시장에서 판매됩니다.
- 이는 다수의 환자를 대상으로도 효과가 있다는 확신이 있어야 하기 때문입니다.
- 어쩌면 컴퓨터 상의 실험으로부터 환자를 대상으로 한 실험으로 바로 넘어갈 수도 있습니다.

![alt text](https://cdn-images-1.medium.com/max/1600/0*-jW_R9CNS3NEjnjx. "Logo Title Text 1")

## 가능성

![alt text](https://294305267s7hqfks2cfh08ip-wpengine.netdna-ssl.com/wp-content/uploads/2017/05/Exscientia-Sanofi-artificial-intelligence-deal-e1494255220247.png "Logo Title Text 1")

- 리드 분자를 찾아내는 초기 단계에는 아무것도 확신할 수 없습니다.
- 실험실에서 행해지는 실험은 여전히 아주 느리고 비싸기 때문에, 가능한 한 리드 분자를 정확하게 찾아내야 합니다. 
- 목표가 암 치료라고 하더라도, 작은 분자의 모든 변형을 실험실에서 확인할 수는 없습니다.
- 7,200만 개는 특정 데이터베이스의 크기일 뿐, 총 분자의 개수는 대략 10⁶⁰ 개에서 10²⁰⁰ 개로 추산됩니다.
- 실험실에서 새로운 분자 하나를 합성하고 시험하는 데에만 수천에서 수만 달러가 들어갑니다. 
- 따라서 초기 추측 단계가 아주, 아주 중요합니다. 
- 머신러닝 모델을 이용하여 필요한 성질을 가졌을 확률이 가장 높은 분자를 찾아낼 수 있습니다.

## 신약 개발 분야에서 머신러닝의 역사

#### 순환 신경망 https://arxiv.org/pdf/1701.01329v1.pdf

![alt text](https://image.slidesharecdn.com/deeplearningbusinessmodelsvnitc-2015-09-13-150912180100-lva1-app6891/95/deep-learning-and-business-models-vnitc-20150913-31-638.jpg?cb=1442162292 "Logo Title Text 1")

- 화학적 언어 모델은 분자와 그 생물학적 활동데이터가 기록된 ChEMBL 데이터베이스로부터 얻은 140만 개의 분자들을 포함한 Smiles 파일에서 학습되었습니다. 
- 그들은 각각 1,024 차원을 가진 3개의 LSTM 층을 사용하여 순환 신경망을 구성하였으며, 각 층 뒤에는 0.2의 dropout 비율을 가진 dropout[65] 층이 신경망을 조정합니다.
- 새로운 분자를 만들어내기 위해서, symbol-by-symbol 모델로부터 50,000,000개의 Smiles symbol이 추출되었습니다.
- 중복된 분자들을 제거한 후, 847,955 개의 새로운 분자를 얻었습니다.

#### 합성곱 신경망 http://arxiv.org/abs/1510.02855v1

![alt text](https://media.springernature.com/lw785/springer-static/image/art%3A10.1186%2Fs12859-017-1702-0/MediaObjects/12859_2017_1702_Fig4_HTML.gif "Logo Title Text 1")

- AtomNet은 최초의 구조 기반 딥 합성곱 신경망으로, 신약 개발을 위한 분자의 대생물 작용을 예측하기 위해 설계되었습니다. 
- 그들은 특징 소재와 위계적 구성의 순환적 개념을 대생물 작용과 화학적 상호작용의 모델링에 적용하는 방법을 설명하였습니다.
- 목표 분자의 구조 정보에 대한 AtomNet의 합성곱 필터의 적용은 이전에 알려진 조절기 없이도 새로운 분자를 성공적으로 예측해 냅니다.
- 신경망의 배치는 입력층과 여러 개의 3차원 합성곱 및 전결합 층, 그리고 약물의 활성도에 확률을 부여하는 로지스틱 비용 층으로 이루어져 있습니다.

#### 생성적 적대 신경망 (GAN)

![alt text](https://cdn-images-1.medium.com/max/1600/1*L_pqMKMNJ8J-sa0Fx0qKpw.jpeg "Logo Title Text 1")

- GAN은 서로 경쟁하는 2개의 신경망으로 구성된 모델로, 2016년에 큰 인기를 끌었습니다.
- 생성자의 목표는 진짜 데이터로 받아들여지는 새로운 객체를 만들어내는 것입니다.
- 판별자는 생성자가 사용하는 요령을 판독하고 진짜 데이터와 생성자가 만들어낸 데이터를 구분해내야 합니다.

![alt text](https://cdn-images-1.medium.com/max/1600/0*MNymJpmxULbHQKVo. "Logo Title Text 1")

- 판별자는 생성자의 가짜 사진을 발견하는 법을 학습하고,
- 생성자는 판별자를 속이기 위한 법을 학습합니다. 

- 조건부 GAN은 사진 화질 향상을 위한 사진 변환 작업에 사용되어 왔습니다. 
- GAN 역시 비교적 "연속적인" 종류의 데이터를 다루는 데 유용합니다.
- 하지만 분자들은 어떨까요? 분자의 구조는 전혀 연속적이지 않고, GAN이 불연속적인 구조를 학습하기란 아주 어렵습니다.
- 여전히, GAN은 새로운 분자를 생성하는 능력을 증명했습니다.

#### 적대적 오토인코더의 적용

![alt text](http://www.inference.vc/content/images/2016/01/Screen-Shot-2016-01-08-at-14-48-25.png "Logo Title Text 1")

- 연구자들은 GAN을 살짝 변형하여 적대적 오토인코더(AAE)라 불리는, 리드 분자를 만들어내기 위한 새로운 모델을 발견했습니다.
- AAE의 핵심은 잠재 표현으로부터 객체를 생성하는 법을 학습하는 것입니다.
- 오토인코더는 입력으로 받아들인 객체와 똑같은 객체를 출력하는 신경망 구조입니다.
- 이 구조에서, 입력 데이터는 잠재 표현(나중에 다시 원래 객체로 되돌리기 위해 입력 데이터로부터 추출한 특징들의 집합)을 학습하는 중간 층을 반드시 통과해야 합니다. 

![alt text](https://cdn-images-1.medium.com/max/1600/0*ovqbOqz_q6FERZCn. "Logo Title Text 1")

- 중간층이 입력 및 출력보다 작거나(낮은 차원이거나), 오토인코더가 특별한 조정 기술을 사용하거나 둘 중 하나이지만, 어떤 경우이든 모든 층에 걸쳐 입력 데이터를 단순히 복사할 수는 없기에 오토인코더는 정말 중요한 정보만을 추출해야 합니다. 
- 그들은 조건부 적대적 오토인코더를 훈련시켜 분자의 지문을 만들어냈으며, 분자에게 요구되는 특성들을 조건으로 사용하였습니다.

![alt text](https://cdn-images-1.medium.com/max/1600/0*KzKdBlERyNuY_RXd. "Logo Title Text 1")

- 판별자는 잠재 표현의 분포를 몇몇 표준 분포들과 구분합니다.
- 만약 잠재 표현의 분포를 표준 분포들과 분간할 수 없게 만든다면, 이 분포에서 추출하여 인코더를 통해 적절한 샘플을 만들어낼 수 있습니다. 
- 또한 이 경우에는 분자에 필요한 성질을 조건으로 사용합니다. 이미 성질이 알려진 분자를 통해 학습하고, 필요한(아마 이전에 본 적 없는) 성질들을 가진 분자를 만들어내야 합니다. 
- 그러면 데이터베이스를 간단히 훑어보는 것만으로도 생성된 분자와 가장 비슷한 성질을 가진 분자를 찾아낼 수 있습니다.
